weighted rating
 FORMULA: WEIGHTED RATING=((numvotes/numvotes + median_votes)*averagerating) + (median_votes/numvotes + median_votes)*Avarage_rating_movies)

```python
C = usable['averagerating'].mean()
m = usable['numvotes'].median()

usable['weighted_rating'] = (
    (usable['numvotes'] / (usable['numvotes'] + m)) * usable['averagerating'] +
    (m / (usable['numvotes'] + m)) * C
)
usable
```

Splitting up movies with multiple genres(splitting the data['genres'])

```python
usable['genres'] = usable['genres'].str.split(',')
usable_split = usable.explode('genres')
usable_split['genre'] = usable_split['genres'].str.strip()
usable_split
```

Calculating mean(weighted_rating)=average weighted rating across all the movies in the genre column.
Calculating sum(numvotes)=total number of votes registered for all movies in the genre column.
Total count of the movies in the genre column(primary_title)

```python
genre_stats = usable_split.groupby('genre').agg({
    'weighted_rating': 'mean',
    'numvotes': 'sum',
    'primary_title': 'count'
}).rename(columns={
    'weighted_rating': 'avg_weighted_rating',
    'numvotes': 'total_votes',
    'primary_title': 'movie_count'
}).reset_index()
genre_stats
```

Ranking the genre and printing them out in order of their perfomance

```python
genre_stats['norm_rating'] = (
    (genre_stats['avg_weighted_rating'] - genre_stats['avg_weighted_rating'].min()) /
    (genre_stats['avg_weighted_rating'].max() - genre_stats['avg_weighted_rating'].min())
)

genre_stats['norm_votes'] = (
    (genre_stats['total_votes'] - genre_stats['total_votes'].min()) /
    (genre_stats['total_votes'].max() - genre_stats['total_votes'].min())
)

genre_stats['performance_score'] = (
    0.7 * genre_stats['norm_rating'] + 0.3 * genre_stats['norm_votes']
)

top_genres = genre_stats.sort_values('performance_score', ascending=False)
print(top_genres[['genre', 'avg_weighted_rating', 'total_votes', 'performance_score']].head(10))
```

VISUALIZATION OF THIS RESULT(BAR PLOT)

```python
top_10_genres = genre_stats.sort_values('performance_score', ascending=False).head(10)
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

barplot = sns.barplot(
    x='performance_score',
    y='genre',
    data=top_10_genres,
    palette='viridis',
    edgecolor='black'
)

plt.title('Top 10 Genres by Performance Score\n(Weighted Rating + Popularity)', fontsize=16, pad=20)
plt.xlabel('Performance Score (0-1 Scale)', fontsize=12)
plt.ylabel('Genre', fontsize=12)

for i, score in enumerate(top_10_genres['performance_score']):
    barplot.text(score + 0.02, i, f'{score:.2f}',
                 va='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

VISUALIZATION OF THIS RESULT(HEAT MAP)

```python
heatmap_data = top_10_genres[['genre', 'avg_weighted_rating', 'total_votes', 'performance_score']]
heatmap_data = heatmap_data.set_index('genre')
heatmap_data = (heatmap_data - heatmap_data.min()) / (heatmap_data.max() - heatmap_data.min())

plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data.T,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    linewidths=0.5,
    cbar_kws={'label': 'Normalized Score'}
)
plt.title('Genre Performance Metrics Comparison', fontsize=16, pad=20)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Metric', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

VISUALIZATION OF THIS RESULT(SCATTER PLOT)

```python
plt.figure(figsize=(12, 8))

scatter = sns.scatterplot(
    x='total_votes',
    y='avg_weighted_rating',
    size='movie_count',
    sizes=(50, 500),
    hue='performance_score',
    data=top_10_genres,
    palette='coolwarm',
    alpha=0.8,
    edgecolor='black'
)
plt.title('Genre Performance: Rating vs. Popularity', fontsize=16, pad=20)
plt.xlabel('Total Votes (Popularity)', fontsize=12)
plt.ylabel('Avg Weighted Rating (Quality)', fontsize=12)
plt.xscale('log')

for line in range(top_10_genres.shape[0]):
    scatter.text(
        top_10_genres['total_votes'].iloc[line] * 1.05,
        top_10_genres['avg_weighted_rating'].iloc[line],
        top_10_genres['genre'].iloc[line],
        horizontalalignment='left',
        size=10,
        color='black'
    )
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

FILLING MISSING ROWS IN THE 'average_gross' COLUMN WITH MEDIAN

```python
filled_movie = merged_info.copy()
filled_movie['average_gross'] = merged_info['average_gross'].fillna(merged_info['average_gross'].median())
filled_movie
```

```python
pure = filled_movie.copy()
pure
```

DROPPING IRRELEVANT COLUMNS

```python
pure_movie_gross = pure.drop(columns=['start_year', 'year', 'foreign_gross', 'domestic_gross', 'studio', 'genre'])
pure_movie_gross
```

DERIVING PROFITABILITY FROM REVENUE IN RELATION TO WEIGHTED RATING TO DETERMINE THE PROFITABILITY FIGURES OF THE GENRES

```python
genres_profitability = pure_movie_gross.groupby('genres').agg({
    'average_gross': 'sum',
    'weighted_rating': 'mean',
    'primary_title': 'count'
}).rename(columns={
    'average_gross': 'total_gross_revenue',
    'weighted_rating': 'avg_rating',
    'primary_title': 'movie_count'
}).sort_values('total_gross_revenue', ascending=False)

genres_profitability['total_gross_revenue'] = genres_profitability['total_gross_revenue'].astype('int64')

pd.set_option('display.float_format', '{:.0f}'.format)
genres_profitability
```

SEABORN PLOT SHOWING THE PROFITABILITY FIGURES AND LEVELS

```python
significant_genres = genres_profitability[genres_profitability['movie_count'] >= 5]

significant_genres = significant_genres.sort_values('total_gross_revenue', ascending=True)

plt.figure(figsize=(12, 8))
sns.barplot(
    data=significant_genres,
    x='total_gross_revenue',
    y=significant_genres.index,
    palette='viridis',
    estimator='sum'
)

plt.title('Total Gross Revenue by Genre (in $)', fontsize=16, pad=20)
plt.xlabel('Total Gross Revenue ($)', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x:,.0f}"))

for i, value in enumerate(significant_genres['total_gross_revenue']):
    plt.text(value, i, f"${value:,.0f}", va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

MATPLOTLIB LOLLIPOP PLOT OF THE ABOVE

```python
significant_genres = genres_profitability[genres_profitability['movie_count'] >= 5]
significant_genres = significant_genres.sort_values('total_gross_revenue', ascending=True)

plt.figure(figsize=(12, 8))
plt.hlines(
    y=significant_genres.index,
    xmin=0,
    xmax=significant_genres['total_gross_revenue'],
    color='skyblue',
    linewidth=2
)

plt.scatter(
    significant_genres['total_gross_revenue'],
    significant_genres.index,
    color='navy',
    s=100,
    alpha=0.8,
    edgecolors='black'
)

plt.title('Total Gross Revenue by Genre (Lollipop Plot)', fontsize=16, pad=20)
plt.xlabel('Total Gross Revenue ($)', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))

for i, value in enumerate(significant_genres['total_gross_revenue']):
    plt.text(
        value + 0.05 * max(significant_genres['total_gross_revenue']),
        i,
        f"${value:,.0f}",
        va='center',
        fontsize=10
    )

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

SHOWING CORRELATION CO-EFFICIENTS BETWEEN REVENUE, RATINGS AND VOTES

```python
correlation_matrix = pure_movie_gross[['average_gross', 'weighted_rating', 'numvotes']].corr()

print("Correlation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Between Revenue, Ratings, and Votes", pad=20)
plt.show()
```
